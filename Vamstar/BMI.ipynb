### Calculating BMI, BMI Category, BMI Range (kg/m2), Health risk

```python
def BMI(df):
    try:
        assert df['WeightKg'] > 0 and df['HeightCm'] > 0
        # checking for correct data. Weight and Height cannot be less than or equal to zero
    except:
        return None
    try:
        x = df['WeightKg']/(df['HeightCm']/100)
    except ZeroDivisionError:
        return None
    
    return x

def BMI_Category(df):
    if df['BMI'] <= 18.4:
        return "Underweight"
    elif df['BMI'] >= 18.5 and df['BMI'] <= 24.9:
        return "Normal weight"
    elif df['BMI'] >= 25 and df['BMI'] <= 29.9:
        return "Over weight"
    elif df['BMI'] >= 30 and df['BMI'] <= 34.9:
        return "Moderately obese"
    elif df['BMI'] >= 35 and df['BMI'] <= 39.9:
        return "Severely obese"
    elif df['BMI'] >= 40:
        return "Very severely obese"
    else:
        return None

def Health_Risk(df):
    if df['BMI'] <= 18.4:
        return "Malnutrition risk"
    elif df['BMI'] >= 18.5 and df['BMI'] <= 24.9:
        return "Low risk"
    elif df['BMI'] >= 25 and df['BMI'] <= 29.9:
        return "Enhanced risk"
    elif df['BMI'] >= 30 and df['BMI'] <= 34.9:
        return "Medium risk"
    elif df['BMI'] >= 35 and df['BMI'] <= 39.9:
        return "High risk"
    elif df['BMI'] >= 40:
        return "Very High risk"
    else:
        return None
    
def BMI_Range(df):
    if df['BMI'] <= 18.4:
        return "18.4 and Below"
    elif df['BMI'] >= 18.5 and df['BMI'] <= 24.9:
        return "18.5 - 24.9"
    elif df['BMI'] >= 25 and df['BMI'] <= 29.9:
        return "25 - 29.9"
    elif df['BMI'] >= 30 and df['BMI'] <= 34.9:
        return "30 - 34.9"
    elif df['BMI'] >= 35 and df['BMI'] <= 39.9:
        return "35 - 39.9"
    elif df['BMI'] >= 40:
        return "40 and above"
    else:
        return None
    
df = df.assign(BMI = df.apply(BMI, axis=1)) 
df = df.assign(BMI_Category = df.apply(BMI_Category, axis=1))
df = df.assign(BMI_Range_kg_m2 = df.apply(BMI_Range, axis=1))
df = df.assign(Health_Risk = df.apply(Health_Risk, axis=1)) 
df
```
